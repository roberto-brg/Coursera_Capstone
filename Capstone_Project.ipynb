# Segmenting and Clustering Neighborhoods in Toronto

## Part 1

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header = 0)
```

```python
for df in dfs:
    print(df.head())
```

```python
df = dfs[0]
df.head()
```

```python
df_cleaned = df[df['Borough'] != 'Not assigned']
df_cleaned.reset_index(drop=True, inplace = True)
df_cleaned
```

```python
r = df_cleaned.groupby('Postal Code').apply(lambda g: g.Neighborhood.value_counts()).reset_index(level=-1) #level=-1
r
```

```python
df2 = r.level_1.groupby(level=0).apply(', '.join).reset_index()
df2
```

```python
df_final = df_cleaned[['Postal Code', 'Borough']].merge(df2[['Postal Code', 'level_1']], on="Postal Code", how = 'inner') #['Postal Code', 'Borough']
df_final = df_final.rename(columns={'level_1': 'Neighborhood'})
df_final
```

```python
df_final.loc[df['Neighborhood'] == 'Not assigned', 'Neighborhood'] = df_final.loc[df['Neighborhood'] == 'Not assigned', 'Borough']
df_final
```
